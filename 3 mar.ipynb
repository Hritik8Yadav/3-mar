```python
#Ans1: The five plots used in seaborn are:
#        Scatterplot: used to identify the relation between two variables of a dataset.
#        catplot: used to plot the dataset variable in categorical way
#        displot: used to plot dataset in the form of histogram
#        barplot: used to plot dataset variables in the form of bars by catogirizing one variable according to second variable.
#        pairplot: used to plot all dataset variables against each other.
```

```python
import seaborn as sb
fmri=sb.load_dataset('fmri')
fmri
```

```python
#Ans2:

sb.lineplot(x=fmri.timepoint,y=fmri.signal,hue=fmri.event,style=fmri.region)
```

```python
titanic=sb.load_dataset('titanic')
titanic
```

```python
sb.boxplot(x='pclass',y='age',data=titanic)
```

```python
sb.boxplot(x='pclass',y='fare',data=titanic)
```

```python
dmd=sb.load_dataset("diamonds")
sb.displot(x='price',data=dmd,hue='cut')
```

```python
iris=sb.load_dataset('iris')
sb.pairplot(data=iris,hue='species')
```
